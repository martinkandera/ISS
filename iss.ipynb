```python
# load the data
# references will be in ref_signals, reference labels in ref_labels, reference count in N_ref.
# tests  will be in test_signals, test labels in test_labels, test count in N_test.
def get_signals(labs):
    signals = []
    N = len(labs)
    for car in labs:
      filename = login + "/" + car + ".wav"
      s, Fs = sf.read(filename)
      signals.append(s)
    return signals, N, Fs

def play_signals(signals, Fs):
  for signal in signals:
    display(Audio(signal, rate=Fs))

files = glob.glob(login + "/*.wav")
names = [re.sub(login + "/", "", s) for s in files]
labels = [re.sub(".wav", "",s) for s in names]
print ("----- test signals ---------")
r = re.compile("^test_");  test_labels = list(filter(r.match, labels))
print (test_labels); test_signals, N_test, Fs = get_signals(test_labels); play_signals (test_signals, Fs)
print ("----- reference signals ---------")
r = re.compile("(?!^test_)");  ref_labels = list(filter(r.match, labels))
print (ref_labels); ref_signals, N_ref, Fs = get_signals(ref_labels); play_signals (ref_signals, Fs)
```

```python
import librosa
import librosa.display

# Funkcia na vykreslenie grafov jedného signálu
def plot_signal_features(signal, sr, title):
    plt.figure(figsize=(12, 8))

    # 1) Waveform
    plt.subplot(3,1,1)
    plt.title(title + " - Waveform")
    plt.plot(signal)
    plt.xlabel("Časové vzorky")
    plt.ylabel("Amplitúda")

    # 2) Spektrogram (STFT)
    # pre výpočet spektrogramu použijeme krátkodobú Fourierovu transformáciu (STFT)
    D = np.abs(librosa.stft(signal, n_fft=1024, hop_length=256))
    # D je frekvencia x frame, prevedieme do dB
    D_db = librosa.amplitude_to_db(D, ref=np.max)

    plt.subplot(3,1,2)
    plt.title(title + " - STFT Spectrogram (dB)")
    librosa.display.specshow(D_db, x_axis='time', y_axis='hz', sr=sr)
    plt.colorbar(format="%+2.0f dB")

    # 3) MFCC
    mfccs = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=13)
    plt.subplot(3,1,3)
    plt.title(title + " - MFCC")
    librosa.display.specshow(mfccs, x_axis='time', sr=sr)
    plt.colorbar()
    plt.tight_layout()
    plt.show()

# Vykreslíme grafy pre všetky testovacie signály
for i, (sig, lbl) in enumerate(zip(test_signals, test_labels)):
    plot_signal_features(sig, Fs, "Test " + lbl)

# Vykreslíme grafy pre všetky referenčné signály
for i, (sig, lbl) in enumerate(zip(ref_signals, ref_labels)):
    plot_signal_features(sig, Fs, "Ref " + lbl)
```
